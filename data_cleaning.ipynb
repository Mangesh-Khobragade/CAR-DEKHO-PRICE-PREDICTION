```python
import pandas as pd


# Function to clean the price string
def clean_price(price_str):
    # Remove the currency symbol and 'Lakh'
    price_str = price_str.replace('', '').replace('Lakh', '').replace(' ','').replace(',', '').replace('Crore', '').strip()
    
    # Return cleaned value
    return float(price_str)

# Apply the cleaning function to the price column
overall_details['price'] = overall_details['price'].apply(clean_price)

# Preview the cleaned data
overall_details.head()
```

```python

# Function to clean the  km string
def clean_km(km_str):
   
    km_str = km_str.replace('km', '').replace(',', '').strip()
    # Return cleaned value
    return int(km_str)

# Apply the cleaning function to the km column
overall_details['km'] = overall_details['km'].apply(clean_km)

# Preview the cleaned data
overall_details.head()
```

```python
overall_details['model year'] = overall_details['modelYear'].astype(str)
```

```python
overall_details.info()
```

```python
import numpy as np
import pandas as pd

# Convert the 'price' column to a Series for easier calculations
price_series = pd.Series(overall_details['price'])

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = price_series.quantile(0.25)
Q3 = price_series.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out values outside of these bounds
# The corrected assignment to the DataFrame
overall_details['filtered_price'] = price_series[(price_series >= lower_bound) & (price_series <= upper_bound)]

print("Filtered overall_details['price']:\n", overall_details['filtered_price'])
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

filtered_price_data = overall_details[['filtered_price']]
min_max_scaler = MinMaxScaler()
overall_details['filtered_price_normalized_min_max'] = min_max_scaler.fit_transform(filtered_price_data)
print(overall_details.head())
```

```python
overall_details = overall_details.drop(columns=['filtered_price_normalized_min_max','price'])
```
